```python
viaggi = [
    {
        "nome": "Viaggio comunitario tra villaggi andini",
        "paese": "Per√π",
        "durata": 18,
        "costo": 2400,
        "attrazioni": [
            "Comunit√† andine rurali",
            "Agricoltura tradizionale",
            "Scambi culturali locali",
            "Turismo comunitario responsabile"
        ],
        "categoria": ["Amici", "Famiglie"],
        "impatto": 9
    },
    {
        "nome": "Tradizioni e sostenibilit√† nel Giappone rurale",
        "paese": "Giappone",
        "durata": 15,
        "costo": 3500,
        "attrazioni": [
            "Villaggi tradizionali",
            "Artigianato locale",
            "Cerimonia del t√®",
            "Stili di vita sostenibili"
        ],
        "categoria": ["Coppie", "Sposi"],
        "impatto": 7
    },
    {
        "nome": "Esperienza rigenerativa nella foresta amazzonica",
        "paese": "Brasile",
        "durata": 14,
        "costo": 2800,
        "attrazioni": [
            "Foresta amazzonica",
            "Progetti di conservazione ambientale",
            "Incontri con comunit√† indigene",
            "Educazione ecologica"
        ],
        "categoria": ["Amici", "Coppie"],
        "impatto": 8
    },
    {
        "nome": "Vita lenta tra risaie e villaggi rurali",
        "paese": "Vietnam",
        "durata": 13,
        "costo": 1900,
        "attrazioni": [
            "Risaie terrazzate",
            "Homestay presso famiglie locali",
            "Cucina tradizionale condivisa",
            "Vita quotidiana rurale"
        ],
        "categoria": ["Famiglie", "Coppie"],
        "impatto": 9
    },
    {
        "nome": "Cammini alternativi sulle Ande peruviane",
        "paese": "Per√π",
        "durata": 21,
        "costo": 2600,
        "attrazioni": [
            "Cammini lontani dal turismo di massa",
            "Comunit√† montane",
            "Spiritualit√† andina",
            "Paesaggi d'alta quota"
        ],
        "categoria": ["Sposi", "Coppie"],
        "impatto": 10
    },
    {
        "nome": "Scambi culturali tra templi e villaggi",
        "paese": "Thailandia",
        "durata": 16,
        "costo": 2100,
        "attrazioni": [
            "Templi buddhisti",
            "Ospitalit√† in villaggi locali",
            "Tradizioni spirituali",
            "Scambi culturali autentici"
        ],
        "categoria": ["Famiglie", "Amici"],
        "impatto": 8
    },
    {
        "nome": "Vita condivisa tra comunit√† rurali",
        "paese": "Colombia",
        "durata": 12,
        "costo": 2000,
        "attrazioni": [
            "Comunit√† rurali",
            "Progetti sociali locali",
            "Caff√® coltivato in modo sostenibile",
            "Incontri con cooperative"
        ],
        "categoria": ["Amici", "Coppie"],
        "impatto": 7
    },
    {
        "nome": "Ritmi lenti e tradizioni nel Giappone montano",
        "paese": "Giappone",
        "durata": 11,
        "costo": 3000,
        "attrazioni": [
            "Villaggi di montagna",
            "Cammini panoramici",
            "Cucina locale stagionale",
            "Incontri con comunit√† locali"
        ],
        "categoria": ["Coppie", "Sposi"],
        "impatto": 9
    },
    {
        "nome": "Esperienza comunitaria tra deserto e oasi",
        "paese": "Marocco",
        "durata": 10,
        "costo": 1800,
        "attrazioni": [
            "Oasi tradizionali",
            "Vita nel deserto",
            "Accoglienza in famiglie locali",
            "Artigianato berbero"
        ],
        "categoria": ["Amici", "Famiglie"],
        "impatto": 8
    },
    {
        "nome": "Turismo rigenerativo tra vulcani e villaggi",
        "paese": "Guatemala",
        "durata": 14,
        "costo": 2200,
        "attrazioni": [
            "Villaggi maya",
            "Tradizioni ancestrali",
            "Paesaggi vulcanici",
            "Progetti di sviluppo comunitario"
        ],
        "categoria": ["Famiglie", "Coppie"],
        "impatto": 9
    }
]



#FUNZIONE VISUALIZZA

def visualizza(viaggi):
    if viaggi:
        print("\nüåç Ecco i viaggi Orizon attualmente disponibili: esperienze pensate per viaggiare con rispetto, incontrare le comunit√† locali e lasciare un impatto positivo.\n")
        for viaggio in viaggi:
            print(f"-"* 63)
            print(f"üìç {viaggio['nome']}")
            print(f"Il viaggio si svolge in {viaggio['paese']} e ha una durata di {viaggio['durata']} giorni.")
            print(f"Ha un costo di {viaggio['costo']} ‚Ç¨ a persona.")
            print(f"Le principali attrazioni turistiche sono:")
            for attrazione in viaggio["attrazioni"]:
                print(f"- {attrazione}")
            print(f"Il viaggio √® ideato per: {', '.join(viaggio['categoria'])}")
            print(f"Ha un impatto positivo sull'ambiente e le comunit√†: {viaggio['impatto']}/10")
            print(f"-"* 63,"\n")
        print("‚ú® Vuoi personalizzare una di queste esperienze? Con Orizon ogni viaggio pu√≤ essere adattato a ritmi, interessi e valori.\n")
    else:
        print("üìå Al momento non ci sono viaggi disponibili. Torna presto: stiamo creando nuove esperienze responsabili e rigenerative.")



#FUNZIONE AGGIUNTA NUOVO ELEMENTO   

def aggiunta(viaggi):
    print("\nüß≠ Creiamo insieme un nuovo viaggio Orizon.")
    print("Inserisci i dettagli: puntiamo su esperienze autentiche, rispettose dell‚Äôambiente e delle comunit√† locali.\n")

    
    #nome
    while True:
        nome = input("üéà Inserisci un nome per il viaggio:").strip()
        
        if nome == "":
            print("\n‚ö†Ô∏è Inserisci un nome, il campo non pu√≤ essere vuoto\n")
        elif any(carattere.isdigit() for carattere in nome):
            print("\n‚ö†Ô∏è Non √® possibile inserire un numero nel nome\n")
        else:
            break

            
    #paese        
    while True:    
        paese = input("üåç Inserisci il paese dove verr√† svolto il viaggio:").strip()
        
        if paese == "":
            print("\n‚ö†Ô∏è Inserisci il Paese, il campo non pu√≤ essere vuoto\n")
        elif any(carattere.isdigit() for carattere in paese):
            print("\n‚ö†Ô∏è Non √® possibile inserire un numero nel paese\n")
        else:
            break
    
    
    #durata
    while True:
        try:
            durata = int(input("‚è≥ Inserisci la durata del viaggio:"))
            if durata > 0:
                break
            print("\n‚ö†Ô∏è La durata deve essere positiva\n")
        except ValueError:
            print("\n‚ö†Ô∏è Inserisci un numero valido\n")

            
    #costo 
    while True:
        try:
            costo= int(input("üí∞ Inserisci il costo del viaggio:"))
            if costo > 0:
                break
            print("\n‚ö†Ô∏è Il costo deve essere positivo\n")
        except ValueError:
            print("\n‚ö†Ô∏è Inserisci un prezzo valido\n")

            
    #non √® stato implementato un controllo se la cella √® vuota appositamente, magari una persona non conosce delle
    #attrazioni per il viaggio.
    attrazioni = [attrazione.strip() for attrazione in input("üé° Inserisci una o pi√π attrazioni:").split(",") 
                  if attrazione.strip() != ""]
    
    
    #categoria
    categorie_accettate = ['Amici', 'Famiglie', 'Coppie', 'Sposi']
    print("üë• A chi √® rivolto questo viaggio? Seleziona una o pi√π categorie.")
    while True:
        categoria_input = input(
            f"Inserisci le categorie del viaggio ({', '.join(categorie_accettate)}): ").strip()

        categoria = [c.strip().capitalize() for c in categoria_input.split(",") if c.strip() != ""]

        if not categoria:
            print("\n‚ö†Ô∏è Devi inserire almeno una categoria.\n")
            continue

        non_valide = [c for c in categoria if c not in categorie_accettate]

        if non_valide:
            print("\n‚ö†Ô∏è Categorie non valide:", ", ".join(non_valide))
            print("Categorie accettate:", ", ".join(categorie_accettate))
            print("‚ö†Ô∏è Riprova.")
            continue

        break

        
    #impatto
    while True:
        try:
            impatto = int(input("üå± Inserisci l'impatto ambientale positivo da 1-10:"))
            if impatto >=1 and impatto <=10:
                break
            print("\n‚ö†Ô∏è Inserisci un numero tra 1 e 10")
        except ValueError:
            print("\n‚ö†Ô∏è Devi inserire un numero")

    print(" ")

    dizio = {'nome':nome,
             'paese':paese,
             'durata':durata,
             'costo':costo,
             'attrazioni':attrazioni,
             'categoria':categoria,
             'impatto':impatto}

    viaggi.append(dizio)
    
    print("\n‚úÖ Viaggio aggiunto con successo al catalogo Orizon! Ecco un riepilogo:\n")
    
    visualizza(viaggi)
    


#FUNZIONE RICERCA SEMPLICE

def ricerca(viaggi):  
    print("\nüîé Troviamo il viaggio Orizon pi√π adatto a te.")
    print("Puoi cercare per Paese, nome, categoria o anche per attrazioni (separa con virgole).\n")

    while True:
        parole = input("Inserisci una o pi√π parole per la ricerca separate da virgola: ").split(",")
        parole = [parola.strip().lower() for parola in parole if parola.strip() != ""]
    
        if not parole:
            print("‚ö†Ô∏è Inserisci almeno una parola chiave.\n")
            continue
        break
    
    print("\n ü™Ñ Ecco cosa abbiamo trovato in base alla tua ricerca:")
    
    risultati = []

    for viaggio in viaggi:
        testo = (
            viaggio["nome"] + " " +
            viaggio["paese"] + " " +
            " ".join(viaggio["categoria"]) + " " +
            " ".join(viaggio["attrazioni"])).lower()

        if any(parola in testo for parola in parole):
            risultati.append(viaggio)

    if not risultati:
        print("\n‚ö†Ô∏è Nessun risultato trovato. Prova con un'altra parola chiave.\n")
    else:
        visualizza(risultati)



#FUNZIONE STATISTICHE

def statistiche(viaggi):
    somma_durata = 0
    somma_costo = 0
    somma_impatto = 0
    conteggio_paesi = {}
    miglior_rapporto = -1
    miglior_viaggio = ""
    costo = None
    impatto = None
    
    if viaggi:
        for viaggio in viaggi:
            somma_durata += viaggio["durata"]
            somma_costo += viaggio["costo"]
            somma_impatto += viaggio["impatto"]
            paese=viaggio["paese"]
            if paese in conteggio_paesi:
                conteggio_paesi[paese]+=1
            else:
                conteggio_paesi[paese]=1
    
            ratio = viaggio["impatto"]/viaggio["costo"]
            if ratio > miglior_rapporto:
                miglior_rapporto = ratio
                miglior_viaggio = viaggio["nome"]
                costo = viaggio["costo"]
                impatto = viaggio["impatto"]
                
    
        media_durata = somma_durata/len(viaggi)
        min_durata,min_nome_durata = min((d["durata"],d["nome"]) for d in viaggi)
        max_durata,max_nome_durata = max((d["durata"],d["nome"]) for d in viaggi)
        
        media_costo = somma_costo/len(viaggi)
        min_costo,min_nome_costo = min((d["costo"],d["nome"]) for d in viaggi)
        max_costo,max_nome_costo = max((d["costo"],d["nome"]) for d in viaggi)
        
        media_impatto = somma_impatto/len(viaggi)
        min_impatto,min_nome_impatto = min((d["impatto"],d["nome"]) for d in viaggi)
        max_impatto,max_nome_impatto = max((d["impatto"],d["nome"]) for d in viaggi)

        print("\nüìä Report Orizon: una panoramica dei nostri viaggi, tra durata, costi e impatto positivo.\n")
        
        print("üåé Distribuzione delle destinazioni.\n")
        print(f"In totale sono presenti {len(viaggi)} viaggi, suddivisi in:")
        
        for paese,numero in conteggio_paesi.items():
            print(f"- {paese}: {numero}")
        
        print("\n‚è≥ Durata delle esperienze\n")
        
        print(f"La media della durata dei viaggi √®: {round(media_durata,2)} giorni")
        print(f'Il viaggio pi√π breve √® "{min_nome_durata}" e dura {min_durata} giorni.')
        print(f'Il viaggio pi√π lungo √® "{max_nome_durata}" e dura {max_durata} giorni.\n')
        
        print("üí∂ Budget indicativo\n")
        
        print(f"La media del costo dei viaggi √®: {round(media_costo,2)} ‚Ç¨ a persona")
        print(f'Il viaggio meno costoso √® "{min_nome_costo}", al prezzo di {min_costo} ‚Ç¨ a persona.')
        print(f'Il viaggio pi√π costoso √® "{max_nome_costo}", al prezzo di {max_costo} ‚Ç¨ a persona.\n')
        
        print("üå± Impatto positivo\n")
        
        print(f"La media dell'impatto positivo sull'ambiente e le comunit√† √®: {round(media_impatto,2)}")
        print(f"Il viaggio con il punteggio pi√π basso di impatto positivo sull'ambiente e le comunit√† √® '{min_nome_impatto}' con {min_impatto}/10.")
        print(f"Il viaggio con il punteggio migliore di impatto positivo sull'ambiente e le comunit√† √® '{max_nome_impatto}' con {max_impatto}/10.\n")

        print("‚öñÔ∏è Valore responsabile: miglior equilibrio tra costo e impatto positivo\n")
        print(f'Il viaggio con il miglior rapporto impatto/costo √® "{miglior_viaggio}". \nImpatto {impatto}/10 | costo {costo} ‚Ç¨ | ratio {round(miglior_rapporto,4)}')
        
    else:
        print("‚ö†Ô∏è Non ci sono ancora viaggi registrati: quando ne aggiungi qualcuno, Orizon potr√† generare il report completo.")

    

#FUNZIONE FILTRAGGIO AVANZATO

def filtraggio_avanzato(viaggi):

    print("\nüß© Filtri avanzati Orizon: combina pi√π criteri per trovare l‚Äôesperienza perfetta pi√π in linea con il tuo modo di viaggiare.")
    print("\n Se un filtro non ti interessa, premi INVIO per saltarlo.\n")

    #paese
    while True:
        paese_raw = input("üåç Paese di destinazione: ").strip()
        if not paese_raw:
            paese = None
            break
        if any(carattere.isdigit() for carattere in paese_raw):
            print("\n‚ö†Ô∏è Il paese non pu√≤ contenere numeri\n")
            continue
        paese = paese_raw.lower()
        break
    
    
    #durata
    while True:
        durata_raw = input("‚è≥ Durata massima del viaggio (in giorni): ").strip()
        if not durata_raw:
            durata_max = None
            break
        if not durata_raw.isdigit():
            print("\n‚ö†Ô∏è Inserisci solo numeri interi positivi.\n")
            continue
        durata_max = int(durata_raw)
        if durata_max <= 0:
            print("\n‚ö†Ô∏è La durata deve essere maggiore di 0.\n")
            continue
        break
    
    
    #costo
    while True:
        costo_raw = input("üí∞ Budget massimo previsto (‚Ç¨): ").strip()
        if not costo_raw:
            costo_max = None
            break
        if not costo_raw.isdigit():
            print("\n‚ö†Ô∏è Inserisci solo numeri interi positivi.\n")
            continue
        costo_max = int(costo_raw)
        if costo_max <= 0:
            print("\n‚ö†Ô∏è Il budget deve essere maggiore di 0.\n")
            continue
        break
    
        
    #impatto
    while True:
        impatto_raw = input("üå± Impatto minimo su ambiente e comunit√† (1‚Äì10): ").strip()
        if not impatto_raw:
            impatto_min = None
            break
        if not impatto_raw.isdigit():
            print("\n‚ö†Ô∏è Inserisci un numero tra 1 e 10.\n")
            continue
        impatto_min = int(impatto_raw)
        if not (impatto_min >=1 and impatto_min <=10):
            print("\n‚ö†Ô∏è Il valore deve essere compreso tra 1 e 10.\n")
            continue
        break
    
    
    #categorie
    categorie_accettate = ['Amici', 'Famiglie', 'Coppie', 'Sposi']
    
    while True:
        categorie_raw = input(f"üë• Categoria di viaggio ({', '.join(categorie_accettate)}), separate da virgola:  ").strip()
        
        if not categorie_raw:
            categorie = None
            break
            
        categorie = [c.strip().capitalize() for c in categorie_raw.split(",") if c.strip() != ""]
    
        if not categorie:
            print("\n‚ö†Ô∏è Inserisci almeno una categoria oppure premi INVIO per saltare.\n")
            continue
    
        non_valide = [c for c in categorie if c not in categorie_accettate]
    
        if non_valide:
                print("\n‚ö†Ô∏è Categorie non valide:", ", ".join(non_valide))
                print("Categorie accettate:", ", ".join(categorie_accettate))
                print("Riprova.\n")
                continue
    
        break

    
    #attrazioni
    attrazioni_raw = input("üé° Attrazioni o esperienze desiderate: ").strip()
    attrazioni = attrazioni_raw.lower() if attrazioni_raw else None
    
    risultati = []
    
    for viaggio in viaggi:
        
        if paese is not None and viaggio["paese"].strip().lower() != paese:
            continue
    
        if durata_max is not None and viaggio["durata"] > durata_max:
            continue
    
        if costo_max is not None and viaggio["costo"] > costo_max:
            continue
    
        if impatto_min is not None and viaggio["impatto"] < impatto_min:
            continue
    
        if categorie:
            viaggio_categorie = [c.strip().lower() for c in viaggio["categoria"]]
            if not any(c.strip().lower() in viaggio_categorie for c in categorie):
                continue
    
        if attrazioni:
            attraz = attrazioni.strip().lower()
            if not any(attraz in a.lower() for a in viaggio["attrazioni"]):
                continue
    
        risultati.append(viaggio)

    if risultati:
        print("\nü™Ñ Ecco cosa abbiamo trovato in base alla tua ricerca:")
        visualizza(risultati)
    else:
        print("\n‚ö†Ô∏è Nessun viaggio trovato con questi filtri.")


        
#MEN√ô

print("Benvenuto in Orizon üåç")

while True:
    print("-" * 40)
    print("Men√π Orizon")
    print("Scegli un‚Äôoperazione: \n")
    print("1Ô∏è‚É£ Visualizza tutti i viaggi")
    print("2Ô∏è‚É£ Aggiungi un nuovo viaggio")
    print("3Ô∏è‚É£ Ricerca semplice")
    print("4Ô∏è‚É£ Filtraggio avanzato")
    print("5Ô∏è‚É£ Statistiche")
    print("0Ô∏è‚É£ Esci")
    print("-" * 40,"\n")
    
    scelta_raw = input("Seleziona un'opzione: ").strip()
    
    if not scelta_raw.isdigit():
            print("\n‚ö†Ô∏è Hai inserito dei caratteri non corretti, inserisci un numero tra 0 e 5.\n")
            continue
        
    scelta = int(scelta_raw)
    if not (scelta >=0 and scelta <=5):
        print("\n‚ö†Ô∏è Il valore deve essere compreso tra 0 e 5. \n")
        continue
    
    if scelta == 1:
        visualizza(viaggi)
        input("Premi INVIO per tornare al men√π...")
    elif scelta == 2:
        aggiunta(viaggi)
        input("Premi INVIO per tornare al men√π...")
    elif scelta == 3:
        ricerca(viaggi)
        input("Premi INVIO per tornare al men√π...")
    elif scelta == 4:
        filtraggio_avanzato(viaggi)
        input("Premi INVIO per tornare al men√π...")
    elif scelta == 5:
        statistiche(viaggi)
        input("Premi INVIO per tornare al men√π...")
    elif scelta == 0:
        print("\nüéâ Grazie per aver scelto Orizon, arrivederci.")
        break


```
